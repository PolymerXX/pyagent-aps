```python
from re import T
from hindsight_pydantic_ai import configure, create_hindsight_tools

configure(
    hindsight_api_url="http://localhost:8888",
    budget="mid",                  # Recall budget: low/mid/high
    max_tokens=4096,               # Max tokens for recall results
    tags=["env:prod"],             # Tags for stored memories
    recall_tags=["scope:global"],  # Tags to filter recall
    recall_tags_match="any",       # Tag match mode: any/all/any_strict/all_strict
)

# Now create tools without passing client — uses global config
Tools = create_hindsight_tools(bank_id="Evan-boy")
```

```python
from hindsight_client import Hindsight
from pydantic_ai import Agent

client = Hindsight(base_url="http://localhost:8888")

agent = Agent(
    "openrouter:xiaomi/mimo-v2-flash",
    tools= Tools,
    instructions= 
    """
    You are a helpful assistant.
    """
)

await agent.run("What do you remember about Evan's preferences?")
```

```python
import asyncio

async def main():
    # First conversation
    r1 = await agent.run(
        "Remember that I prefer functional programming patterns "
        "and I'm building a data pipeline in Python."
    )
    print(r1.output)

    # Later conversation -- agent recalls context
    r2 = await agent.run("What approach should I take for error handling?")
    print(r2.output)

await main()
```
